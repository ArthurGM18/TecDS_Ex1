### Lendo a base de dados do Yahoo Finance sobre fornecimento dados do mercado de ações.
### Pegando dados entre 10/09/2019 e 09/10/2019.

```python
df = pdr.DataReader('GE', 'yahoo', start='2019-09-10', end='2019-10-09')
```

### Faz o cálculo do retorno simples e retorno logarítmo

```python
df['simple return'] = (df['Adj Close']/df['Adj Close'].shift(1))-1
df['log return'] = np.log(df['Adj Close']/df['Adj Close'].shift(1))
df.head()
```
